## Чтение и парсинг данных

```python
data = open('../file2.txt').readlines()
```

```python
walks = []
rot_x = []
rot_y = []
rot_z = []

acc_x = []
acc_y = []
acc_z = []

prev_walk = False
for row in data[1:]:
    values = row.split(' ')
    if len(values) == 6:
        walks.append(prev_walk)
    
        rot_x.append(float(values[0]))
        rot_y.append(float(values[1]))
        rot_z.append(float(values[2]))
        acc_x.append(float(values[3]))
        acc_y.append(float(values[4]))
        acc_z.append(float(values[5][:-1]))
        
    else:
        walks.append(values[0] == "true")
        prev_walk = values[0] == "true"
#         print(prex)
    
        rot_x.append(float(values[1]))
        rot_y.append(float(values[2]))
        rot_z.append(float(values[3]))
        acc_x.append(float(values[4]))
        acc_y.append(float(values[5]))
        acc_z.append(float(values[6][:-1]))
```

```python
import pandas as pd
```

## Подсчет фичей для классификации

```python
T = 10
```

```python
def mean(data):
    return sum(data) / float(len(data))
```

```python
d = {
    'target': [], 
    'avg_rot_x': [], 'avg_rot_y': [], 'avg_rot_z': [],
    'min_rot_x': [], 'min_rot_y': [], 'min_rot_z': [],
    'max_rot_x': [], 'max_rot_y': [], 'max_rot_z': [],
    'avg_acc_x': [], 'avg_acc_y': [], 'avg_acc_z': [],
    'min_acc_x': [], 'min_acc_y': [], 'min_acc_z': [],
    'max_acc_x': [], 'max_acc_y': [], 'max_acc_z': [],
}

for i in range(10, len(walks)):
    d['target'].append(walks[i])
    
    d['avg_rot_x'].append(mean(rot_x[i - 10:i]))
    d['avg_rot_y'].append(mean(rot_y[i - 10:i]))
    d['avg_rot_z'].append(mean(rot_z[i - 10:i]))
    d['min_rot_x'].append(min(rot_x[i - 10:i]))
    d['min_rot_y'].append(min(rot_y[i - 10:i]))
    d['min_rot_z'].append(min(rot_z[i - 10:i]))
    d['max_rot_x'].append(max(rot_x[i - 10:i]))
    d['max_rot_y'].append(max(rot_y[i - 10:i]))
    d['max_rot_z'].append(max(rot_z[i - 10:i]))
    
    d['avg_acc_x'].append(mean(acc_x[i - 10:i]))
    d['avg_acc_y'].append(mean(acc_y[i - 10:i]))
    d['avg_acc_z'].append(mean(acc_z[i - 10:i]))
    d['min_acc_x'].append(min(acc_x[i - 10:i]))
    d['min_acc_y'].append(min(acc_y[i - 10:i]))
    d['min_acc_z'].append(min(acc_z[i - 10:i]))
    d['max_acc_x'].append(max(acc_x[i - 10:i]))
    d['max_acc_y'].append(max(acc_y[i - 10:i]))
    d['max_acc_z'].append(max(acc_z[i - 10:i]))
```

```python
df = pd.DataFrame(data=d)
```

```python
df.head()
```

```python
X = df.drop(['target'], axis=1)
```

```python
y = df['target'].astype(int)
```

## Классификация текущего действия как хотьбы

```python
from catboost import CatBoostClassifier
```

```python
model = CatBoostClassifier()
```

```python
model.fit(X, y, verbose=True)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(model.predict(X), y) 
```

## Сохранение модели в coreml
